```python
import numpy as np
import matplotlib.pyplot as plt
import math
from tqdm import tqdm, trange
import warnings
warnings.simplefilter('ignore')
```

```python
g_beta_plus = 1.0
g_alpha_plus = 0.5
g_beta_minus = 1.2
g_alpha_minus = 0.8
prob = 0.5
```

```python
def f_g_plus(x, g_beta_plus, g_alpha_plus):
    f_of_x = ((g_beta_plus**g_alpha_plus)/math.gamma(g_alpha_plus))*(x**(g_alpha_plus-1))*(np.exp(-g_beta_plus*x))
    return f_of_x
```

```python
def f_g_minus(x, g_beta_minus,g_alpha_minus):
    f_of_x = ((g_beta_minus**g_alpha_minus)/math.gamma(g_alpha_minus))*((-x)**(g_alpha_minus-1))*(np.exp(g_beta_minus*x))
    return f_of_x
```

```python
x_plus = np.linspace(0, 2, num=100)
x_minus = np.linspace(-2, 0, num=100)
```

```python
fig = plt.figure()
plt.plot(x_plus, f_g_plus(x_plus, g_beta_plus, g_alpha_plus), label='plus')
plt.plot(x_minus, f_g_minus(x_minus, g_beta_minus, g_alpha_minus), label='minus')
# plt.axhline(0, color='black')
plt.axvline(0, color='black',linewidth=0.4)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc = 'upper right')
plt.show()
```

```python
fig = plt.figure()
plt.plot(x_plus, np.log(f_g_plus(x_plus, g_beta_plus, g_alpha_plus)), label='plus')
plt.plot(x_minus,np.log(f_g_minus(x_minus, g_beta_minus, g_alpha_minus)) , label='minus')
# plt.axhline(0, color='black')
plt.axvline(0, color='black',linewidth=0.4)
plt.xlabel('x')
plt.ylabel('lnf(x)')
plt.legend(loc = 'upper right')
plt.show()
```
